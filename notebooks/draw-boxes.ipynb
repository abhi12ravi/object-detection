```python
def convert_bbox(coco_bbox):
    x_top_left=coco_bbox[0]
    y_top_left= coco_bbox[1]
    width = coco_bbox[2]
    height = coco_bbox[3]
    
    x_min = x_top_left
    y_min = y_top_left
    x_max = x_top_left + width
    y_max = y_top_left + height

    bbox = [x_min, y_min, x_max, y_max]
    
    return bbox
```

```python
bboxes = [convert_bbox([846, 145, 146, 477]),
 convert_bbox([848, 216, 175, 551]),
convert_bbox([74, 159, 75, 81]),
convert_bbox([153, 124, 658, 643])]

img = bbv.draw_multiple_rectangles(img, bboxes)
labels = ["Person", "Person", "Car", "Car"]
bbv.add_multiple_labels(img, labels, bboxes)
imgplot = plt.imshow(img)
plt.show()
```
